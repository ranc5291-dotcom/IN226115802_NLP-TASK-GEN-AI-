Task 1: Conceptual Understanding

Q1: What is the difference between "Love" and "love" in NLP?
Answer:
In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.
"Love" → may represent a proper noun (e.g., name)
"love" → represents the emotion/action
Without lowercasing, the model treats them as separate words, increasing vocabulary size unnecessarily.

Q4: Explain the difference between stemming and lemmatization
Answer:
Stemming: Cuts words to root form (may not be meaningful)
Example: "running" → "run" / "runn"
Lemmatization: Converts words into meaningful base form using vocabulary
Example: "running" → "run"

```python
import re
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Lowercase
    text = text.lower()
    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize
    tokens = text.split()
    # Remove short words except 'no', 'not'
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]
    cleaned_sentence = " ".join(tokens)
    return tokens, cleaned_sentence
```

```python
texts = [
    "I have 2 dogs!!!",
    "This is good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now"
]
for t in texts:
    print(preprocess_text(t))
```

Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"

    ]
results = []
for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("\n------------------------------")
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    results.append((sentence, tokens, clean))
```

Task 4: Token Analytics

```python
import numpy as np
analytics = []
for original, tokens, clean in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    analytics.append({
        "sentence": original,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_length": avg_length
    })
    print("\nSentence:", original)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", avg_length)
```

Task 5: Frequency Analysis

```python
from collections import Counter
all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)
counter = Counter(all_tokens)
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))
print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences }
# Run
output = full_pipeline(test_sentences)
print(output)
```
